```python
import pandas as pd
import numpy as np

df=pd.read_csv("icd10_Oct_7_participant.csv")
df.head()
```

```python
data=df.copy(deep=True)

data["p41270"]=data["p41270"].map(str)+"|"+data["p20002_i0"].map(str)

data.drop("p20002_i0",axis=1,inplace=True)

data["flag"]=np.zeros(502410)

data
```

```python
k=0
droplist=[]
for s in data["p41270"]:
    if(s=="nan|nan"):
        droplist.append(k)
        print(k)
    k+=1
```

```python
data.drop(droplist,axis=0,inplace=True)
```

```python
data.reset_index(inplace=True)
```

```python
data.drop("index",axis=1,inplace=True)
data
```

```python
k=0
for s in data["p41270"]:
    flag=False
    temp=s.split("|")
    for j in temp:
        if((j.find("I50")!=-1)|(j.find("I11.0")!=-1)|(j.find("I13.0")!=-1)|(j.find("I13.2")!=-1)|(j.find("heart failure")!=-1)):
            data["flag"][k]=1
            flag=True
    if(flag==True):
        print(k)
    k+=1
```

```python
data.to_csv("icd10_Oct_7_flag.csv")
```
